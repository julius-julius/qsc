```python
def runit(state,gloop,listParams,debugmode):
    cutP,QaiShift,errorTollerance,maxIterations,cutQai = listParams;
    #global p
    ######################
    def scan(s,listOut):
        iniS,resS,resDelta,dDelta,itr,shifterror,atime,newtonGoal,flagError,flagSaved = listOut
        if 'dDelta||' in s:
            dDelta = float((s.split('||'))[1]);
        if 'resDelta||' in s:
            resDelta = float((s.split('||'))[1]);
        if 'iniS||' in s:
            iniS = float((s.split('||'))[1]);
        if 'resS||' in s:
            resS = float((s.split('||'))[1]);
        if 'atime||' in s:
            atime = float((s.split('||'))[1]);
        if 'itr||' in s:
            itr = int((s.split('||'))[1]);
        if 'shifterror||' in s:
            shifterror = float((s.split('||'))[1]); 
        if 'newtonGoal||' in s:
            newtonGoal = float((s.split('||'))[1]); 
        if 'flagError||' in s:
            flagError = int((s.split('||'))[1]); 
        if 'flagSaved||' in s:
            flagSaved = int((s.split('||'))[1]); 
        return [iniS,resS,resDelta,dDelta,itr,shifterror,atime,newtonGoal,flagError,flagSaved]
    ######################
    iniS = 0.;
    resS = 0.;
    dDelta = 0.;
    resDelta = 0.;
    itr = 0;
    shifterror = 0.;
    newtonGoal = 0.;
    flagError = 1;
    flagSaved = 0;
    atime = 0.
    listOut = [iniS,resS,resDelta,dDelta,itr,shifterror,atime,newtonGoal,flagError,flagSaved]
    #listOut = list(range(8))
    #precs.append([icur,gloop-0.0,dgloop-0.0])
    p = subprocess.Popen(["wolframscript","-script","TypeI_script.wls"
                          ,str(state[1])#nb[0]
                          ,str(state[2])#nb[0]
                          ,str(state[3])#nf[3]
                          ,str(state[4])#nf[2]
                          ,str(state[5])#nf[2]
                          ,str(state[6])#nf[1]
                          ,str(state[7])#na[0]
                          ,str(state[8])#na[0]
                          ,str(state[9])#sol
                          ,str(cutP)#cutP
                          ,str(cutP+2)#nPoints
                          ,str(cutQai)#cutQai
                          ,str(QaiShift)#QaiShift
                          ,"30"#epsilon
                          ,"186"#WP
                          ,str(gloop)#gloop
                          ,("1" if gloop<1/10 else "10")#intrepolation order
                          ,("1" if gloop<1/10 else "0")#intrepolation method
                          ,str(maxIterations) #number of iterations
                          ,str(errorTollerance)#error tolerence
                         ], stdout=subprocess.PIPE)
    for line in iter(p.stdout.readline, b''):
        topr = (line).decode('UTF-8');
        if debugmode == 1:
            print(topr, end='');
        listOut = scan(topr,listOut);
    p.stdout.close()
    p.wait()
    if debugmode == 1:
        clear_output(wait=True)
    return listOut    
```

```python
# determining which way of boosting the shift precision is better
def BoostShift(state,gloop,listParams,debugmode):
    cutP0, QaiShift0, errorTollerance0, maxIterations0, cutQai0= listParams;
    dQaiShift = 10;
    dcutQai = 4;
    listOut0 = runit(state,gloop,[cutP0,QaiShift0,errorTollerance0,0,cutQai0],debugmode)
    listOut1 = runit(state,gloop,[cutP0,QaiShift0+dQaiShift,errorTollerance0,0,cutQai0],debugmode)
    listOut2 = runit(state,gloop,[cutP0,QaiShift0,errorTollerance0,0,cutQai0+dcutQai],debugmode)
    t0,p0=[listOut0[i] for i in [6,5]]
    t1,p1=[listOut1[i] for i in [6,5]]
    t2,p2=[listOut2[i] for i in [6,5]]
    rate1 = (p0-p1)/(t1-t0)
    rate2 = (p0-p2)/(t2-t0)
    if rate1 > rate2:
        print("increase shift: ",rate1," vs ",rate2," getting: ", p0,">",p1)
        res = [QaiShift0 + dQaiShift, cutQai0]
    else:
        print("increase cutQai: ",rate2," vs ",rate1," getting: ", p0,">", p2)
        res = [QaiShift0, cutQai0 + dcutQai]
    return res
```
